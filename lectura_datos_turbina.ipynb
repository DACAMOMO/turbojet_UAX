```python
datos_arrancada = np.loadtxt('arrancada_36000.ASC', comments='#')
#t[s] P0 [Pa]	P2 [Pa]	P3t [Pa]	P4t [Pa]	P5t [Pa]	P6t [Pa]	T2t [K]	T3t [K]	T4t1 [K]	T4t2 [K]	T5t1 [K]	T5t2 [K]	T8t [K]	Empuje [N]	fuel [kg/s]	aire [kg/s]	/s]
tiempo = datos_arrancada[:,0]
p3t = datos_arrancada[:,3]
E = datos_arrancada[:,14]
T4t1 = datos_arrancada[:,9]
T4t2 = datos_arrancada[:,10]
c = datos_arrancada[:,15]
G = datos_arrancada[:,16]
```

```python
plt.figure()
plt.plot(tiempo,p3t)
plt.xlabel('tiempo [s]')
plt.ylabel('p3t')
plt.show()
```

```python
plt.figure()
plt.plot(tiempo,T4t1,'x-r',label='T4t1')
plt.plot(tiempo,T4t2,'+-b',label='T4t2')
plt.xlabel('tiempo [s]')
plt.ylabel('T4t [K]')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(tiempo,c,'x-r',label='c')
plt.ylabel('Gasto [kg/s]')
plt.legend()
plt.subplot(2,1,2)
plt.plot(tiempo,G,'+-b',label='G')
plt.xlabel('tiempo [s]')
plt.ylabel('Gasto [kg/s]')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(tiempo,E)
plt.xlabel('tiempo [s]')
plt.ylabel('E [N]')
plt.show()
```
